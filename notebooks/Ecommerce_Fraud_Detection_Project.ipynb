```python
import pandas as pd

chunk_size = 100000

first_chunk = True

for file in [
    "Fraudulent_E-Commerce_Transaction_Data.csv",
    "Fraudulent_E-Commerce_Transaction_Data_2.csv"
]:
    for chunk in pd.read_csv(file, chunksize=chunk_size):
        chunk.to_csv(
            "Fraudulent_E-Commerce_Combined.csv",
            mode="w" if first_chunk else "a",
            header=first_chunk,
            index=False
        )
        first_chunk = False

print("Merge completed!")
```

```python
Q1 = df["Transaction Amount"].quantile(0.25)
Q3 = df["Transaction Amount"].quantile(0.75)

print("Q1 =", Q1)
print("Q3 =", Q3)
```

```python
IQR = Q3 - Q1

print("IQR =", IQR)
```

```python
Lower_Bound = Q1 - (1.5 * IQR)
Upper_Bound = Q3 + (1.5 * IQR)

print("Lower Bound =", Lower_Bound)
print("Upper Bound =", Upper_Bound)
```

```python
df["High Value Transaction"] = df["Transaction Amount"] > Upper_Bound
```

```python
df["High Value Transaction"].value_counts()
```
